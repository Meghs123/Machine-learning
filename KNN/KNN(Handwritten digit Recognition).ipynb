```python
# I want to find prediction for query points.
# Start with KNN Code.
# Step 1: Find K Nearest Neighbours , we will find these based on euclids distance.
```

```python
# Distance formula to find distance between 2 numpy arrays - eucliedean distance
def distance(x1,x2):
    return np.sqrt(sum(x1-x2)**2)
```

```python
# KNN algorithm
def knn(x,y,querypoint,k=5):
    # Pick K nearest neighbour
    vals=[]
    # For every point for x i m going to comp[ut the distance
    for i in range(x.shape[0]):
        
        #Compute distance
        d=distance(querypoint, x[i]) # Compute distance between querypoint and x.
        vals.append((d,y[i]))     # Add distance to vals and append y class.
    print(vals)    #Show the distace from the predicting point and class.
    
    # Step-2 , sort the array and find K Nearest Points.
    vals=sorted(vals)
    vals=vals[:k]  # From starting to k points.

    #Majority vote
    vals=np.array(vals)
    new_values=np.unique(vals[:,1], return_counts=True)  # Vals contain distance and y.And here, we get unique values for y.
    
    # Index of the maximum count
    index=new_values[1].argmax()  #argmax function will the index.
    
    # Map this index with my data.
    pred=new_values[0][index]
    
    return pred

    
```
